This notebook will take a simulated sorghum dataset (via directory of .ply files), and based on the set parameters output a single vs paired leaf dataset. 

Some statistics about the output dataset will then be visualized.


```python
# Configuration:

# --- IO ---

# Where is the sorghum processed dataset file in h5py format?
input_dataset_h5py_path = '/space/ariyanzarei/sorghum_segmentation/dataset/2022-03-10/sorghum__labeled_train.hdf5'

# Where should the generated dataset be output?
output_dataset_directory_path = f'/space/murph186/sorghum_segmentation/dataset/2022-03-10_leaf_c'

# Should generated data be persisted?
persist_output = True

# --- Generation parameters ---

# How many points should each generated sample contain
down_sample_n = 80

# What is the maximum distance bewteen leaf centers to be considered for
# inclusion into generated dataset as pair?
centroid_dist_threshold = 0.1 

# How many total samples (before split) should be generated?
total_sample_n = 8000
```

```python
leaf_dataset = generate_leaf_dataset(
  input_dataset_h5py_path, 
  down_sample_n=down_sample_n,
  centroid_dist_threshold=centroid_dist_threshold,
  total_sample_n=total_sample_n
  )

print(f"[Info] Successfully generated {len(leaf_dataset['points'])} leaves")

(train_dataset, validation_dataset, test_dataset) = split_dataset(leaf_dataset)
```

```python
if persist_output:

  print(f"[Info] Saving results to {output_dataset_directory_path}")
  if not os.path.exists(output_dataset_directory_path):
    os.makedirs(output_dataset_directory_path)

  output_to_h5py(train_dataset, os.path.join(
    output_dataset_directory_path, 'sorghum_leaf__labeled_train.hdf5'))
  output_to_h5py(validation_dataset, os.path.join(
    output_dataset_directory_path, 'sorghum_leaf__labeled_validation.hdf5'))
  output_to_h5py(test_dataset, os.path.join(
    output_dataset_directory_path, 'sorghum_leaf__labeled_test.hdf5'))
```
