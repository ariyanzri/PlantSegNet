```python
import tensorflow as tf

train_event_file = "/space/murph186/sorghum_segmentation/models/model_checkpoints/leaf_net_2022_03_10_v1/lightning_logs/version_1/events.out.tfevents.1659557864.gauss.58483.0" 

train_step_loss = []
train_step_acc = []

val_loss = []
val_acc = []

for e in tf.compat.v1.train.summary_iterator(train_event_file):
  for v in e.summary.value:

    if v.tag == 'train_leaf_loss_step':
      train_step_loss.append(v.simple_value)
    
    if v.tag == 'val_leaf_acc':
      val_acc.append(v.simple_value)

    if v.tag == 'val_leaf_loss':
      val_loss.append(v.simple_value)
```

```python
import matplotlib.pyplot as plt
ax = plt.axes()
ax.plot(range(len(val_acc)), val_acc)
ax.set(xlim=(0, 20), 
       xlabel='epoch', 
       ylabel='percent accuracy',
       title='Validation Accuracy');
```

```python
ax = plt.axes()
ax.plot(range(len(val_loss)), val_loss)
ax.set(xlim=(0, 20), 
       xlabel='epoch', 
       ylabel='loss',
       title='Validation Loss');
```
