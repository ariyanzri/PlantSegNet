```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append("..")
from data.utils import distinct_colors
import random
```

```python
data = np.loadtxt("/space/ariyanzarei/sorghum_segmentation/dataset/archive/Pheno4D/Tomato05/T05_0325_a.txt", dtype=float)
```

```python
print(data.shape)
downsample_indexes = random.sample(
    np.arange(0, data.shape[0]).tolist(),
    min(10000, data.shape[0]),
)
downsampled_data = data[downsample_indexes]
```

```python
d_colors = distinct_colors(len(list(set(downsampled_data[:,3]))))
colors = np.zeros((downsampled_data[:,3].shape[0],3))
for i, l in enumerate(list(set(downsampled_data[:,3]))):
    colors[downsampled_data[:,3] == l, :] = d_colors[i]
    
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(projection='3d')
ax.scatter(downsampled_data[:,0],downsampled_data[:,1],downsampled_data[:,2], s=3, c=colors)
plt.show()
```
