```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append("..")
from data.utils import distinct_colors
```

```python
data = np.loadtxt("/space/ariyanzarei/sorghum_segmentation/testdata.txt", dtype=float)
```

```python
d_colors = distinct_colors(len(list(set(data[:,3]))))
colors = np.zeros((data[:,3].shape[0],3))
for i, l in enumerate(list(set(data[:,3]))):
    colors[data[:,3] == l, :] = d_colors[i]
    
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(projection='3d')
ax.scatter(data[:,0],data[:,1],data[:,2], s=3, c=colors)
plt.show()
```
